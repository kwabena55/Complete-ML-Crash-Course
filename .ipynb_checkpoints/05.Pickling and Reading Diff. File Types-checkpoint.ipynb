# Convert Python Dictionary into a JSON and Read

```python
cars = [
    {"Name":"Honda", "Price": 10000, "Model":2005, "Power": 1300},
    {"Name":"Toyota", "Price": 12000, "Model":2010, "Power": 1600},
    {"Name":"Audi", "Price": 25000, "Model":2017, "Power": 1800},
    {"Name":"Ford", "Price": 28000, "Model":2009, "Power": 1200},
         
]
```

```python
# Write the python dictg as Json 

import json

with open('Data/cars.json', 'w') as f:
    json.dump(cars, f) 
```

```python
#Read json file
pd.read_json('Data/cars.json')
```
