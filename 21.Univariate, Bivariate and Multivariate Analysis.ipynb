Univariate,Bivariate,Multivariate
* Leverage More Histograms in Univariate Analysis
* Pairplot in Multivariate
* Bar graphs, scatter plot in Bivariate Analysis

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

```

```python
df=pd.read_csv("https://raw.githubusercontent.com/uiuc-cse/data-fa14/gh-pages/data/iris.csv")
```

```python
df.head()
```

Multivariate Analysis

```python
sns.pairplot(df.corr())
```

```python
sns.heatmap(df.corr(),annot=True)
```

```python
sns.pairplot(df,hue="species")
```

Univariate Analysis
* Either generate invidual labels or use hue in Seaborn to help visualuze 

```python
setosa=df["species"]=="setosa"
```

```python
df[setosa]
```

```python
sns.pairplot(df[setosa])
```

```python
df["sepal_length"].plot(kind="hist", bins=5)
```

```python
#this plot will show the range of the sepal length for a paritcular species
plt.plot(df[setosa]["sepal_length"], np.zeros_like(df[setosa]["sepal_length"]),'o',color="white")
plt.show()
```

```python
versicolor=df["species"]=="versicolor"
virginica=df["species"]=="virginica"
```

```python
plt.plot(df[versicolor]["sepal_length"], np.zeros_like(df[versicolor]["sepal_length"]),'o',color="yellow")
plt.show()
```

```python
plt.plot(df[virginica]["sepal_length"], np.zeros_like(df[virginica]["sepal_length"]),'o',color="green")
plt.show()
```

```python
plt.figure(figsize=(13,8))
plt.plot(df[setosa]["sepal_length"], np.zeros_like(df[setosa]["sepal_length"]),'o',color="white")
plt.plot(df[versicolor]["sepal_length"], np.zeros_like(df[versicolor]["sepal_length"]),'o',color="yellow")
plt.plot(df[virginica]["sepal_length"], np.zeros_like(df[virginica]["sepal_length"]),'o',color="green")
plt.show()
```

Bivariate Analysis
* Mostly you can write a for loop to plot all features or derive insights from pairplot. Also leverage seaborns facegrid plot

```python
sns.FacetGrid(df,hue="species", size=5).map(plt.scatter,"sepal_length","sepal_width")
plt.legend()
plt.show()
```
