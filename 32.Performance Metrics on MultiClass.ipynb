Performance Metrics on MultiClass problems

```python
from sklearn.metrics import classification_report,confusion_matrix
```

```python
c='cat'
d='dog'
f='fox'
```

```python
y_true=[c,c,c,c,c,c,f,f,f,f,f,f,f,f,f,f,d,d,d,d,d,d,d,d,d]
```

```python
y_pred=[c,c,c,d,f,f,f,f,f,f,f,d,d,d,f,f,d,d,d,d,c,c,d,d,d]
```

```python
import seaborn as sns
cm=confusion_matrix
sns.heatmap(cm(y_true,y_pred),annot=True)
```

```python
print(classification_report(y_true,y_pred))
```
