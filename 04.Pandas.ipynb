# Pandas
* Python programming language for data manipulation and analysis. In particular, it offers data structures and operations for manipulating numerical tables and time series

```python
import pandas as pd
import numpy as np
```

```python
#Generate a DF from numpy array
df= pd.DataFrame(np.arange(20).reshape(5,4),index=['row1','row2','row3','row4','row5'], columns=['a', 'b', 'c', 'd'])
```

```python
df
```

# Slicing elements in a Dataframe
* loc ... accessing row 
* iloc... accesing both rows and columns

```python
df.loc["row1"]
```

```python
type(df.loc["row1"])
```

```python
df.loc["row1"][0:2]
```

```python
df.iloc[1:3,0:4]
```

```python
type(df.iloc[1:3,0:4])
```
